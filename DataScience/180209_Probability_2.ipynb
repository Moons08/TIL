### Probability Model (확률 모형)

우리가 다루는 데이터는 실험, 혹은 조사를 통해 얻어진다. 이 데이터들은 그 이면에 어떤 특성에 따라 생성되며, 그 특성을 알기 위해 우리는 데이터를 분석한다.

데이터 분석에 확률론을 적용하면 데이터를 생성한 구체적인 데이터 생성기가 존재하지 않더라도 다음처럼 가상의 데이터 생성기가 존재한다고 생각할 수 있다.

1. 데이터를 뽑을 수 있는 후보자 집합 : 가능한 모든 데이터 값으로 이루어진 **표본 공간**이 존재한다고 가정.
1. 데이터를 뽑는 방법 : 표본 공간의 모든 사건에 대해서 확률이 배정되어 있다고 가정한다. 

이처럼 (1)표본 공간과 (2)확률이 정해져 있으면 이 두 가지를 사용하여 데이터를 생성할 수 있다. 이를 **확률 모형**이라고 부른다.

### Sampling, Realization

우리가 가진 데이터가 확률 모형이라고 하는 가상의 주사위에 의해 생성된 것이라고 할 때, 이 주사위를 던져 데이터를 생성하는 과정을 **sampling** 또는 **realization** 이라고 한다.

샘플링은 다른 의미로도 사용되는데 많은 수의 데이터 집합에서 일부 데이터만 선택하는 과정도 샘플링이라고 한다.

### 데이터의 특성

확률 모형과 실제 데이터는 다음과 같은 관계가 있다.
> - 확률 모형으로부터 데이터를 여러번 생성하는 경우, 실제 데이터 값은 매번 달라질 수 있지만 **확률 모형 자체는 변하지 않는다.**


> - 확률 모형은 우리가 직접 관찰할 수 없다. 다만 확률 모형에서 만들어지는 실제 데이터 값을 이용하여 확률 모형이 이러한 것일 거라고 추정하고 가정할 뿐이다.


> - 확률 모형에서 만들어지는 실제 데이터의 값은 확률 모형이 가진 특성을 반영하고 있다. 다만, **데이터의 갯수가 적을 수록 부정확**하여 확률 모형의 특징을 정확하게 추정할 수 없다.

### 데이터 분석 과정

1. 데이터 확보
1. 확보된 데이터를 어떤 확률 모형의 표본으로 가정
1. 데이터의 특성으로부터 확률 모형의 특성을 추정
1. 구해진 확률 모형의 특성으로 해당 확률 모형의 종류를 결정, 모수 추정
1. 구해진 확률 모형으로부터 다음에 생성될 데이터나 특성을 예측

# 확률 변수

확률 변수는 표본 공간의 모든 표본에 대해 어떤 실수 값을 붙인 것이다. 따라서 확률 변수는 표본 공간을 정의역(domain)으로 가지고 실수를 공역(range)으로 가지는 함수라고 할 수 있다.

확률과 확률 변수의 차이점
- 확률은 표본으로 이루어진 집합. 사건에 대해 할당된 숫자
- 확률 변수는 표본 하나 하나에 대해 할당된 숫자


- 확률은 0부터 1 사이의 숫자만 할당 가능
- 확률 변수는 모든 실수 범위의 숫자를 할당 가능

**확률 변수를 사용하면** 모든 표본은 하나의 실수 숫자로 변하기 때문에 **표
본 공간을 실수의 집합 즉, 수직선으로 표시할 수 있다.**

예를 들어, $a$보다 같거나 크고 $b$보다 작은 숫자의 집합인 사건 $A$는 다음과 같은 기호로 표시한다.

$$A = \{ \omega ; a \leq X(\omega) < b \} = \{ a \leq X < b \}$$

## discrete random variable (이산 확률 변수)

확률 변수 값이 연속적이지 않고 떨어져 있는 경우 ex) 주사위

- 주의: 이산 확률 변수의 정의는 값의 이산성, 가능한 경우가 유한하다는 말이 아님. 이산 확률 변수도 연속 확률 변수와 같이 무한대의 가능한 경우가 있을 수 있다.

확률 변수를 정의한다는 것은 표본이라는 추상적인 개념대신 숫자라는 명확한 개념을 사용하겠다는 의미이다. 계산이 가능한 것은 숫자 뿐이므로 분석을 위해서는 결국 표본의 특성(feature)을 숫자로 변환하는 단계가 필요하다.

### continuous random variable (연속 확률 변수)

확률 변수 값이 연속적, 무한개인 경우 ex) 시각을 가리키는 시계바늘

시계바늘이 특정한 시각을 가리킬 확률은 0도이다. (나누는 값이 무한대이기 때문) 따라서 확률을 정의할 때 특정한 각도가 아닌 구간(interval)이라는 사건(event)에 대해 확률을 정의하였다.

확률 변수를 사용할 경우 이와 같은 연속 확률 분포의 사건을 수직선 상의 실수 구간으로 변환할 수 있다.
