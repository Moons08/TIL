```python
import time
from selenium import webdriver
from PIL import Image as pil
```

```python
driver = webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=Aqx41JrNTSw")
time.sleep(10)
driver.save_screenshot('screenshot.png') # 스샷 저장
```

```python
pil_im = pil.open('screenshot.png') # 저장된 스샷 불러오기
print(pil_im.size) 
pil_im
```

```python
pil_im.thumbnail((512,512)) # pillow 로 사이즈 변환
pil_im.save("screenshot_thumnail.png")
pil.open('screenshot_thumnail.png')
```

```python
element = driver.find_element_by_css_selector('#movie_player') #영상사이즈 측정
print(type(element))
location = element.location
size = element.size
location, size
```

```python
pil_im = pil.open('screenshot.png')
left, top = location['x'], location['y']
right, bottom = left + size['width'], top + size['height']
area = (left, top, right, bottom)
area
```

```python
pil_im = pil_im.crop(area)
```

```python
pil_im.save('screenshot_element.png')
pil.open('screenshot_element.png')
```
