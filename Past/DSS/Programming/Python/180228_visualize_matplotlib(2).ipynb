### bar plot

```python
y = np.random.randint(10, size=5)
x = range(5)
plt.bar(x, y)
plt.show()
```

yerr, xerr 로 에러 바 추가 

```python
error = np.random.randint(1, 5, size=5)
plt.bar(x, y, yerr=error, align='center', alpha=0.7)  # alpha : 투명도 지정
plt.show()
```

`plt.barh` : horizontal bar plot

```python
plt.barh(x, y, xerr=error, align='center', alpha=0.6)
plt.show()
```

#### 바 플롯 두개를 겹쳐서 표현

```python
y1 = np.random.randint(5, 10, size=5)
y2 = np.random.randint(1, 5, size=5)
print(y1, y2)
x = np.arange(5)
plt.bar(x, y1)
plt.bar(x, y2)
plt.tight_layout()
plt.show()
```

같은 데이터를 다른방식으로 한꺼번에 표현
- x축과 align(bar의 위치)를 조정

```python
bar_width, opacity = 0.4, 0.6 #opacity 투명도

plt.bar(x, 
            y1, bar_width, alpha=opacity, 
            color='b', label='randint_1', align="edge")
plt.bar(x + bar_width, 
            y2, bar_width, alpha=opacity, 
            color='r', label='randint_2', align="edge") 

plt.xlabel('X Label') # X 라벨
plt.ylabel('Value')   # Y 라벨
plt.xticks(x + bar_width, ('A', 'B', 'C', 'D', 'E')) # xtick 추가
plt.title('Random Integer') # 제목추가
plt.legend() # 범례추가
plt.tight_layout()
plt.show()
```

### stem plot
bar plot과 유사, 폭이 없음
- 이산 확률 함수, 자기상관관계 묘사에 유용

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.stem(x, np.cos(x), '-.')
plt.show()
```

### pie plot
카테고리 별 값의 상대적 비교

```python
data = np.random.randint(1,6, size=5)
labels = ['A','B','C','D','E']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red']
explode = (0, 0, 0.1, 0, 0.1)
plt.pie(data, 
        labels=labels, 
        colors=colors, 
        explode=explode, # 강조에 유용
        shadow=True,
        startangle=30,
        autopct='%1.2f%%',
       ) # 소숫자리 수 표현, g 는 반올림
plt.axis('equal') # 원 형태 유지 - 없으면 찌그러짐
plt.tight_layout()
plt.show()
```

### scatter plot
2차원 데이터 집합의 상관 관계 파악에 유용

```python
# uniform (균등분포)
x1 = np.random.rand(500)
y1 = np.random.rand(500)
plt.subplot(211)
plt.scatter(x1, y1)

# gaussian (정규분포)
x2 = np.random.randn(500)
y2 = np.random.randn(500)
plt.subplot(212)
plt.scatter(x2, y2)

plt.tight_layout()
plt.show()
```

bubble chart
- 데이터가 3, 혹은 4차원일 경우 점의 크기, 색을 이용한다

ex) 부동산데이터 - x 집값, y1 역과의 거리, y2 집 면적, y3 집 형태(아파트, 오피스텔)

```python
#4차원 데이터의 경우 - x, y, colors, area
x = np.random.randint(1,10, size=20)
y1 = np.random.randint(1,10, size=20)
y2 = np.pi * np.arange(10,50,2) ** 2
y3 = np.random.rand(20)
plt.scatter(x, y1, s=y2, c=y3, alpha=0.5) # s 넓이, c 색
plt.show()
```

### histogram
`hist` 명령은 `bins`인수로 데이터를 집계할 구간 정보를 받는다.

```python
x = np.random.randn(1000)
arrays, bins, patches = plt.hist(x, bins=10)
# bins = 10 -> 10개의 구간으로 잘라 해당되는 수를 카운트
print(arrays) # 구간당 카운트
print(bins) # 구간
print(patches)
plt.show()
```

```python
x1 = np.random.rand(1000)
plt.subplot(211)
plt.hist(x1, bins=50, normed=True)

x2 = np.random.randn(1000)
plt.subplot(212)
plt.hist(x2, bins=200, normed=True) # 잘게 쪼개서 카운트

plt.tight_layout()
plt.show()
```

### Imshow
image 데이터처럼 행과 열을 가진 행렬 형태의 2차원 데이터 분석에 유용

```python
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.images[0]
X
```

```python
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
methods = [
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
    'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
    'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'
]
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

### Contour 
입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우 -> 3차원 자료

입력 변수 x, y는 `meshgrid` 명령으로 그리드 포인트 행렬화


### 3D surface: Axe3D

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
```

```python
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black', linewidths=1.8)
plt.show()
```

```python
# Axes3D로 표현하면 다음과 같다.
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```

```python
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y) # contour와 마찬가지로 그리드포인트로 바꿔준다
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```

```python
# contour로 표현
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black', linewidths=1.8)
plt.show()
```
