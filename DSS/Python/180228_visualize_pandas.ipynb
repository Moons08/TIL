pandas series, DataFrame에 `plot` 메서드로 시각화 가능

```python
ts = pd.Series(np.random.randn(1000), 
               index=pd.date_range('1/1/2000', 
                                   periods=1000))
ts = ts.cumsum() #Return cumulative sum over requested axis.
ts.plot()
plt.show()
```

`plot`메서드의  `kind`인수를 바꾸면 여러 플롯을 그릴 수 있다.
- bar, pie, hist, kde, box, scatter, area

```python
iris = sns.load_dataset("iris")
```

```python
iris.tail()
```

```python
iris[:5].plot.bar(rot=0) 
#rot = Rotation for ticks (xticks for vertical, yticks for horizontal plots)
plt.ylim(0, 7)
plt.tight_layout()
plt.show()
```

```python
iris.plot.scatter("sepal_width", "petal_width") 
plt.tight_layout()
plt.show()
```

```python
iris.plot.box() 
plt.tight_layout()
plt.show()
```
