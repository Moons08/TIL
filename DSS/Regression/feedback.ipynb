> 추가로 설명하고 싶은 부분  
 - 그래프 형태에 따라 'Censored'데이터 형태가 발견될 수도 - 여러가지 이유로. ex) 없어진 상태에서 왔거나, 미니멈으로 다 채워줬다거나
 이런 데이터들은 기본적으로 따로 빼서 처리를 해야함. 어떻게 해서 잘될지는 실제로 해봐야함. 
 대부분의 팀이 interaction 을 쓰지 않음. 변수의 개수가 굉장히 많아진다는 단점이 있지만, 중요한 것들, 동시에 영향을 미칠 수 있을 만한 것들, 
 택시의 경우 시간과 요일에 대한 것들이 복합적으로 영향을 미칠 수 있다면 사용할 수 있음. (아무데나 사용하면 안됨. 변수의 개수가 폭팔적으로 증가)
> 
 - 잔차의 정규성을 띄고 안띄고가 중요한게 아님. 정규성을 안띄는데 어떤 모양으로 안띄는가. 거기서 내가 어떤 규칙을 추가적으로 찾아낼 수 있는가
 거기서 도움을 받으려고, 잔차의 정규성 플랏을 보는 것. Box-cos 변환이 있음. 종속변수를 정규분포로 만드려고 하는 이유는 OLS가 잘되는 이유가 많기 때문. 
 박스 코스 변환은 람다라는 파라미터를 가진 변환하는 것. 람다가 1/2이면 로그와 같은 등, 수치를 바꿔가면서 종속변수의 분산 폭을 조절할 수 있음.
 로그를 했을 때 상수값에 대해서 결과값이 달라짐. 최소한 로그 안의 형태가 음수가 나와서는 안되기 때문에 아무렇게나 조절할 순 없음. 
 사실 최종 퍼포먼스에서는 큰 영향을 끼치지 않지만, 굳이 만들고 싶다면 사용할 수 있음. 
> 
 - OLS가 잘했다 못했다는 잔차의 정규성인게 중요한 것이 아님. 테스트 퍼포먼스만이 유의미. 힌트를 얻으려고 하는 것이기 때문에. 
> 
 - 시간이 거리가 비례하는데, 그게 평일과 주말과 같은가. 실수값에 대해서는 interaction을 잘 해주지 않음. 실수와 실수가 곱해져서 상관관계를 가지지 않는 것이 많음. 로그 취하면 더하기로 떨어지기 때문에. 보통 인털액션은 카테고리, 카테고리끼리 클래스별로 슬래프를 다르게 만들거나.
