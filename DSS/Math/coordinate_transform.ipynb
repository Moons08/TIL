```python
import scipy.misc
%matplotlib inline
```

```python
f = sp.misc.face(gray=True)
plt.imshow(f, cmap=mpl.cm.bone)
plt.grid(False)
plt.show()
```

```python
g_1 = np.array([1, 1])/np.sqrt(2)
g_2 = np.array([-1, 1])/np.sqrt(2)
```

```python
A = np.vstack([g_1, g_2]).T
A
```

```python
import scipy.ndimage

f1 = sp.ndimage.affine_transform(f, A)
plt.imshow(f1, cmap=mpl.cm.bone)
plt.grid(False)
plt.show()
```

다음 기저 벡터를 이용하여 이미지를 변환하라. 파이썬으로 변환한 이미지를 만들기 전에 어떤 이미지가 나올지 생각해 보자.

$$g_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix} ,\;\;
g_2 = \begin{bmatrix} -0.5 \\ 1 \end{bmatrix}$$

```python
g_1 = np.array([1, 0])
g_2 = np.array([-0.5, 1])
```

```python
A = np.vstack([g_1, g_2]).T
A
```

```python
import scipy.ndimage

f1 = sp.ndimage.affine_transform(f, A)
plt.imshow(f1, cmap=mpl.cm.bone)
plt.grid(False)
plt.show()
```
