```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from scipy import stats

# set data
boston = load_boston()
X = boston.data
y = boston.target
```

```python
# modeling
lm = LinearRegression()
lm.fit(X,y)
params = np.append(lm.intercept_,lm.coef_)
predictions = lm.predict(X)

# 바이어스 오그멘테이션
X_a = np.append(np.ones((len(X),1)), X, axis=1)

# Mean Squared Error : RSS / (N - K)
MSE = (sum((y-predictions)**2))/(len(X_a)-len(X_a[0]))

# when wanna use 'DafaFrame'
# X_a = pd.DataFrame({"Constant":np.ones(len(X))}).join(pd.DataFrame(X)) 
# MSE = (sum((y-predictions)**2))/(len(X_a)-len(X_a.columns))

var_b = MSE*(np.linalg.inv(np.dot(X_a.T,X_a)).diagonal())
sd_b = np.sqrt(var_b)
ts_b = params/ sd_b

p_values =[2*(1-stats.t.cdf(np.abs(i),(len(X_a)-1))) for i in ts_b]

sd_b = np.round(sd_b,3)
ts_b = np.round(ts_b,3)
p_values = np.round(p_values,3)
params = np.round(params,4)

result = pd.DataFrame()
result["Coefficients"],result["Standard Errors"],result["t values"],result["P values"] = [params,sd_b,ts_b,p_values]
print(result)
```
